# Data Visualization

```python
titanic = sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(4, 3))
sns.countplot(x='survived', data=titanic,color='red')
plt.title('Survival Count')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()

# Bar chart of survival counts
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# Survival bar graph as per gender ratios
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=titanic, x='age', bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

#Histogram of age distributions
```

```python
sns.heatmap(titanic.isna(),cmap='viridis',cbar=True)
plt.title("Missing Data Heatmap")
plt.show()

# Heatmap of missing data column-wise
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Distribution of Passenger Fare by Ticket Class')
plt.xlabel('Passenger Class (1st, 2nd, 3rd)')
plt.ylabel('Fare (USD)')
plt.show()

# Box plot showing  distribution of passenger fare by ticket class
```
